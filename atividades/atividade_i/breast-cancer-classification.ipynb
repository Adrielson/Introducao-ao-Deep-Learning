```python
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(20, 12))

# Draw the heatmap
sns.heatmap(corr, mask=mask, linewidths=1, annot=True, fmt=".2f", cmap="coolwarm")

# Show the plot
plt.show()
```

```python
corr_matrix = numeric_df.corr().abs()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

tri_df = corr_matrix.mask(mask)

to_drop = [x for x in tri_df.columns if any(tri_df[x] > 0.92)]

df = df.drop(to_drop, axis=1)

print(f"O data frame reduzido tem {df.shape[1]} colunas.")
```

```python
X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# Encode the labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Codificar os rótulos (diagnósticos malignos ou benignos) como 0 e 1
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

# MLP criação e treinamento
Multilayer Perceptron (MLP)

```python
# Definir o modelo MLP
mlp_model = Sequential()

# Adicionar uma camada de entrada explícita
mlp_model.add(Input(shape=(X_train.shape[1],)))

# Primeira camada oculta (64 neurônios, ativação relu)
mlp_model.add(Dense(64, activation='relu'))

# Segunda camada oculta (32 neurônios, ativação relu)
mlp_model.add(Dense(32, activation='relu'))

# Camada de saída (classificação binária, ativação sigmoid)
mlp_model.add(Dense(1, activation='sigmoid'))

# Compilar o modelo com otimizador Adam e função de perda para classificação binária
mlp_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Usar early stopping (parada antecipada) para evitar overfitting
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
# Treinar o modelo MLP
mlp_history = mlp_model.fit(X_train, y_train_encoded, batch_size=32, epochs=100, 
                            validation_data=(X_test, y_test_encoded), callbacks=[early_stop])
```

```python
# Avaliar o MLP
mlp_loss, mlp_accuracy = mlp_model.evaluate(X_test, y_test_encoded)

# Fazer previsões com o MLP (com threshold de 0.5 para a ativação sigmoid)
y_pred_mlp = (mlp_model.predict(X_test) > 0.5).astype("int32")
mlp_class_report = classification_report(y_test_encoded, y_pred_mlp)

# Calcular as métricas
accuracy_mlp = accuracy_score(y_test_encoded, y_pred_mlp)
precision_mlp = precision_score(y_test_encoded, y_pred_mlp)
recall_mlp = recall_score(y_test_encoded, y_pred_mlp)
f1_mlp = f1_score(y_test_encoded, y_pred_mlp)
roc_auc_mlp = roc_auc_score(y_test_encoded, y_pred_mlp)

# Exibir as métricas
print("Accuracy_mlp:", accuracy_mlp)
print("Precision_mlp:", precision_mlp)
print("Recall_mlp:", recall_mlp)
print("F1 Score_mlp:", f1_mlp)
print("ROC AUC_mlp:", roc_auc_mlp)
```

```python
cm = confusion_matrix(y_test_encoded, y_pred_mlp)
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.title('Matriz de Confusão')
plt.xlabel('Classes Previstas')
plt.ylabel('Classes verdadeiras')
plt.show()
```

#  Decision Tree


```python
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(dt_model, filled=True, feature_names=X.columns, class_names=label_encoder.classes_)
plt.show()
```

```python
y_pred_dt = dt_model.predict(X_test)

accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy_dt: {accuracy_dt}")
print(classification_report(y_test, y_pred_dt))
```

```python
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt, pos_label='M')  # Considerando 1 como a classe positiva
recall_dt = recall_score(y_test, y_pred_dt, pos_label='M')
f1_dt = f1_score(y_test, y_pred_dt, pos_label='M')
y_pred_proba_dt = dt_model.predict_proba(X_test)[:, 1]  # Pegando a probabilidade da classe 1
roc_auc_dt = roc_auc_score(y_test, y_pred_proba_dt)


print("Accuracy_dt:", accuracy_dt)
print("Precision_dt:", precision_dt)
print("Recall_dt:", recall_dt)
print("F1 Score_dt:", f1_dt)
print("ROC AUC:_dt", roc_auc_dt)
```

```python
cm = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.title('Matriz de Confusão')
plt.xlabel('Classes Previstas')
plt.ylabel('Classes verdadeiras')
plt.show()
```
