```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, ReLU, GlobalAveragePooling2D, Dense, Add, AveragePooling2D, Flatten, MaxPooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.metrics import SparseCategoricalAccuracy
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from scipy.stats import shapiro, f_oneway, kruskal
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Concatenate
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Carregue o conjunto de dados CIFAR-10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Gaussian Normalization
X_train = (X_train - np.mean(X_train)) / np.std(X_train)
X_test = (X_test - np.mean(X_test)) / np.std(X_test)

# Defina o número de execuções (neste caso, 10)
num_execucoes = 10

# Inicialize listas para armazenar métricas de desempenho
val_accuracy_scores_model1 = []
val_accuracy_scores_model2 = []
```

```python
# Carregue o conjunto de dados CIFAR-10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Especifique a porcentagem do dataset que você deseja manter (por exemplo, 10%)
percentage_to_keep = 0.5

# Calcule o número de amostras a serem mantidas
num_samples_to_keep = int(percentage_to_keep * len(X_train))

# Reduza o dataset de treinamento e os rótulos correspondentes
X_train = X_train[:num_samples_to_keep]
y_train = y_train[:num_samples_to_keep]

# Aplique a Gaussian Normalization aos dados reduzidos
X_train = (X_train - np.mean(X_train)) / np.std(X_train)

# Você também pode fazer o mesmo para o conjunto de teste, se desejar
num_samples_to_keep_test = int(percentage_to_keep * len(X_test))
X_test = X_test[:num_samples_to_keep_test]
y_test = y_test[:num_samples_to_keep_test]

# Aplique a Gaussian Normalization aos dados reduzidos do conjunto de teste
X_test = (X_test - np.mean(X_test)) / np.std(X_test)

# Agora, X_train e X_test contêm 10% do tamanho original do conjunto de treinamento e teste,
# respectivamente, e ambos foram normalizados usando a média e o desvio padrão do conjunto de treinamento reduzido.
# Defina o número de execuções (neste caso, 10)
num_execucoes = 10

# Inicialize listas para armazenar métricas de desempenho
val_accuracy_scores_model1 = []
val_accuracy_scores_model2 = []
```

# **DenseNet121**

Arquitetura de rede neural convolucional (CNN) profunda amplamente utilizada em tarefas de visão computacional, como classificação de imagens.

A principal característica das redes DenseNet, incluindo o DenseNet-121, é a **conexão densa entre camadas**.

Em uma rede densa, cada camada recebe conexões diretas das camadas anteriores e também envia suas ativações para todas as camadas subsequentes. Isso é diferente das arquiteturas convencionais, onde as camadas são conectadas sequencialmente.

# **Arquitetura DenseNet121**
**Camada de Entrada:**
A entrada é uma imagem colorida de formato (32, 32, 3), o que significa que tem 32 pixels de altura, 32 pixels de largura e 3 canais de cores (vermelho, verde e azul).

**Camada Inicial de Convolução:**
Esta camada realiza uma convolução inicial com 64 filtros de tamanho (7, 7) e com uma taxa de stride de (2, 2) para reduzir a resolução da imagem pela metade.
Em seguida, aplica normalização por lotes (BatchNormalization) e ativação ReLU (Rectified Linear Unit).

**Camada de Max Pooling:**
Uma camada de max-pooling com tamanho de janela (3, 3) e uma taxa de stride de (2, 2) é aplicada após a primeira convolução. Isso reduz ainda mais a resolução da imagem.

**Blocos Densos:**
A arquitetura DenseNet é caracterizada pela presença de vários blocos densos, cada um consistindo em camadas de convolução, normalização por lotes e ativação ReLU.
Foi defnido quatro blocos densos, que são construídos usando funções auxiliares dense_block.
Cada bloco denso é composto por camadas de convolução com 32 filtros de tamanho (3, 3) e ativação ReLU. As ativações dessas camadas são concatenadas com as ativações de camadas anteriores, permitindo que a rede acesse ativamente informações de camadas anteriores.

**Camadas de Transição:**
Entre os blocos densos, camadas de transição são inseridas para reduzir a dimensão espacial da saída.
As camadas de transição incluem normalização por lotes, ativação ReLU, uma convolução com fator de compressão (0,5) e uma camada de max-pooling com tamanho de janela (2, 2) para reduzir a resolução.

**Camada de Global Average Pooling:**
Após os blocos densos, a rede aplica uma camada de Global Average Pooling. Isso calcula a média das ativações em todos os pixels da última camada convolucional, resultando em uma saída de tamanho fixo.

**Camada de Saída:**
A camada de saída é uma camada totalmente conectada (Dense) com 10 unidades, correspondendo às 10 classes no conjunto de dados CIFAR-10.
A função de ativação usada na camada de saída é softmax, que converte as saídas em probabilidades de pertencer a cada classe.

```python
# Loop externo para executar o modelo 10 vezes
for _ in range(num_execucoes):
    # Divida aleatoriamente o conjunto de dados em treinamento, validação e teste
    X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.3, random_state=None)

    # Defina o número de épocas de paciência desejado
    paciente = 2

    # Defina a arquitetura do modelo 1
    model1 = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=1),
        tf.keras.layers.Conv2D(64, kernel_size=(3, 3), padding='valid', activation='relu'),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=1),
        tf.keras.layers.Conv2D(128, kernel_size=(3, 3), padding='valid', activation='relu'),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=1),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(200, activation='relu'),
        tf.keras.layers.Dense(10, activation="softmax")
    ])

    # Compile o modelo
    model1.compile(loss='SparseCategoricalCrossentropy', optimizer='adam', metrics=['accuracy'])

    # Configure o callback EarlyStopping para o modelo 1
    early_stopping_model1 = EarlyStopping(monitor='val_loss', patience=paciente, restore_best_weights=True)

    # Treinamento do modelo 1 com acompanhamento de métricas de validação
    history_model1 = model1.fit(X_train_split, y_train_split, epochs=2, batch_size=64, validation_data=(X_val_split, y_val_split))

    # Armazene as métricas de validação do modelo 1
    val_accuracy_scores_model1.append(history_model1.history['val_accuracy'])

    # Plote as curvas de loss e accuracy para o modelo 1
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history_model1.history['loss'], label='Training Loss')
    plt.plot(history_model1.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Model 1 Loss')

    plt.subplot(1, 2, 2)
    plt.plot(history_model1.history['accuracy'], label='Training Accuracy')
    plt.plot(history_model1.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Model 1 Accuracy')

    plt.show()

# Certifique-se de que seus rótulos estejam no formato one-hot para o modelo 2
    num_classes = 10
    y_train_split_one_hot = tf.keras.utils.to_categorical(y_train_split, num_classes)
    y_val_split_one_hot = tf.keras.utils.to_categorical(y_val_split, num_classes)

    # Defina a arquitetura do modelo 2
    def dense_block(x, num_layers, growth_rate):
        for _ in range(num_layers):
            # BatchNormalization and ReLU
            x = BatchNormalization()(x)
            x = ReLU()(x)
            # Convolution
            x = Conv2D(growth_rate, (3, 3), padding='same')(x)
            # Concatenate with previous layers
            x = Concatenate()([x, x])
        return x

    def transition_block(x, compression_factor):
        num_filters = int(x.shape[-1])
        x = BatchNormalization()(x)
        x = ReLU()(x)
        x = Conv2D(int(num_filters * compression_factor), (1, 1))(x)
        x = AveragePooling2D((2, 2))(x)
        return x

    def densenet121(input_shape, num_classes):
        input = Input(shape=input_shape)

        # Initial Convolution
        x = Conv2D(64, (7, 7), strides=(2, 2), padding='same')(input)
        x = BatchNormalization()(x)
        x = ReLU()(x)
        x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

        # Dense Blocks
        x = dense_block(x, 6, 32)
        x = transition_block(x, 0.5)
        x = dense_block(x, 12, 32)
        x = transition_block(x, 0.5)
        x = dense_block(x, 24, 32)
        x = transition_block(x, 0.5)
        x = dense_block(x, 16, 32)

        # Global Average Pooling and Classification Layer
        x = GlobalAveragePooling2D()(x)
        x = Dense(num_classes, activation='softmax')(x)

        model = Model(inputs=input, outputs=x)

        return model

    # Define input shape and number of classes
    input_shape = (32, 32, 3)  # Adjust input shape as needed
    num_classes = 10  # Adjust the number of classes

    # Create the DenseNet-121 model
    model2 = densenet121(input_shape, num_classes)

    # Compile the model and specify optimizer, loss, and metrics
    model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Configure o callback EarlyStopping para o modelo 2
    early_stopping_model2 = EarlyStopping(monitor='val_loss', patience=paciente, restore_best_weights=True)

    # Treinamento do modelo 2 com acompanhamento de métricas de validação
    history_model2 = model2.fit(X_train_split, y_train_split_one_hot, epochs=2, batch_size=64, validation_data=(X_val_split, y_val_split_one_hot))

    # Armazene as métricas de validação do modelo 2
    val_accuracy_scores_model2.append(history_model2.history['val_accuracy'])


    # Plote as curvas de loss e accuracy para o modelo 2
    import matplotlib.pyplot as plt

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history_model2.history['loss'], label='Training Loss')
    plt.plot(history_model2.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Model 2 Loss')

    plt.subplot(1, 2, 2)
    plt.plot(history_model2.history['accuracy'], label='Training Accuracy')
    plt.plot(history_model2.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Model 2 Accuracy')

    plt.show()
```

```python
# Realize os testes de normalidade (Shapiro-Wilk) para ambas as listas de métricas
shapiro_stat_model1, shapiro_p_model1 = shapiro(val_accuracy_scores_model1)
shapiro_stat_model2, shapiro_p_model2 = shapiro(val_accuracy_scores_model2)

# Verifique se as amostras seguem uma distribuição normal
alfa = 0.05
if shapiro_p_model1 > alfa and shapiro_p_model2 > alfa:
    print('Todos os grupos seguem uma distribuição normal.')

    # Realize o teste ANOVA para comparar os grupos
    statistic, p_value = f_oneway(*val_accuracy_scores_model1, *val_accuracy_scores_model2)
    print('Por meio da aplicação do teste ANOVA, verificou-se que:')
    print(f'statistic: {statistic:.4f}')
    print(f'p_value: {p_value:.4f}')
    print('Sendo assim:')
    if p_value < alfa:
        print('Os grupos possuem diferenças estatisticamente significativas.')
    else:
        print('Os grupos não possuem diferenças estatisticamente significativas.')
else:
    print('Algum dos grupos não segue uma distribuição normal.')

    # Realize o teste de Kruskal-Wallis para comparar os grupos
    statistic, p_value = kruskal(*val_accuracy_scores_model1, *val_accuracy_scores_model2)
    print('Por meio do teste de Kruskal-Wallis, verificou-se que:')
    print(f'statistic: {statistic:.4f}')
    print(f'p_value: {p_value:.4f}')
    print('Portanto:')
    if p_value < alfa:
        print('Os grupos possuem diferenças estatisticamente significativas.')
    else:
        print('Os grupos não possuem diferenças estatisticamente significativas.')
```
